```python
from qiskit import IBMQ, QuantumCircuit, transpile, assemble, execute
from qiskit.tools.monitor import job_monitor

# Load IBMQ account
IBMQ.load_account()
provider = IBMQ.get_provider(hub='ibm-q')
backend = provider.get_backend('ibm_montreal')

# Create a quantum circuit
qc = QuantumCircuit(2, 2)
qc.h(0)
qc.cx(0, 1)
qc.measure([0, 1], [0, 1])

# Transpile the circuit for the ibm_montreal backend
transpiled_circuit = transpile(qc, backend=backend)

# Execute the circuit on the ibm_montreal backend
job = execute(transpiled_circuit, backend=backend)

# Monitor the job status
job_monitor(job)

# Get the results
result = job.result()
counts = result.get_counts(qc)
print(counts)
```
