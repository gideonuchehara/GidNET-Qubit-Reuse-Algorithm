## 关于量噪

**量噪 (QEP, Quantum Error Processing)** 是由百度研究院[量子计算研究所](https://quantum.baidu.com)研发的量子噪声处理工具集，旨在从软件层面处理近期量子设备中广泛存在的量子噪声。目前，该工具集提供四类强大的量子噪声处理功能：量子性能评估、量子噪声刻画、量子噪声缓释、以及量子纠错。其中：

+ **量子性能评估**用于高效评估量子计算机的性能，衡量量子计算机可扩展性。QEP 提供标准随机基准测试、交替式基准测试、交叉熵基准测试、酉噪声随机基准测试、直接保真度估计、跨平台保真度估计等工具。

+ **量子噪声刻画**用于完整刻画量子计算机的量子噪声，精确标定量子态和量子门信息。QEP 提供量子态层析、量子过程层析、量子探测器层析、量子门集层析、量子谱层析等工具。

+ **量子噪声缓释**用于显著提升量子计算机的计算结果精度，消除量子门和量子测量的噪声，扩展量子设备的计算能力。QEP 提供量子门零噪声外插、量子测量噪声缓释、截断 Neumann 序列、局域噪声模型等多种量子噪声缓释方法。

+ **量子纠错**用于保护由于环境噪声和硬件缺陷引起的量子信息错误，是容错量子计算的必要工具。QEP 提供稳定码模拟器，可以模拟基于稳定码理论构建的量子纠错码，帮助用户研究各种类型噪声对纠错码性能的影响、评估不同纠错码的性能、并评估量子算法对错误的鲁棒性。

## 常见问题

**Q: 使用量噪需要安装哪些依赖项?**

**A:** 量噪是基于量易伏 [QCompute](https://quantum-hub.baidu.com/opensource) 开源 SDK 开发，所以在使用量噪之前需要预先安装 QCompute。当用户安装量噪时会自动下载安装这个关键依赖包，其他所需的依赖项如下:

```bash
matplotlib>=3.5.2
networkx>=2.8.3
numpy>=1.21.3
pandas>=1.4.3
qiskit>=0.37.0
qiskit_ibmq_provider>=0.19.2
scipy>=1.7.1
seaborn>=0.11.2
tqdm>=4.64.0
qcompute>=3.3.0
qutip>=4.7.0
```

为了更好的用户体验，我们建议在使用量噪前将依赖项更新到最新版本。
