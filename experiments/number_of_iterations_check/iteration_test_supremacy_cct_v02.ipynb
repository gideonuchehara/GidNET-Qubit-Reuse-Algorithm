```python
def score_elements(elements):
    widths = [ele[0] for ele in elements]
    min_width =  min(widths)
    scores = []
    for element in elements:
        length_min_circuit, probability = element
        # score = probability / length_min_circuit  # You can adjust this scoring formula as needed
        # score = 1 / (length_min_circuit * (1 / probability))
        score = probability * (min_width/length_min_circuit)  # You can adjust this scoring formula as needed
        scores.append(score)
    return scores
```

```python
# circuit_sizes = ["4x4", "4x5", "5x5", "5x6", "6x6", "6x7", "7x7", "7x8", "8x8", "8x9",
#                 "9x9", "9x10", "10x10", "10x11", "11x11", "11x12", "12x12"]
# circuit_sizes = ["4x4", "4x5", "5x5", "5x6", "6x6", "6x7"]
circuit_sizes = ["10x11", "11x11", "11x12", "12x12"]

directory_path = "/home/gideonuchehara/Projects/Circuit_Cutting_Qubit_Reuse/GidNET-qubit-reuse-algorithm/Notebook/experiments/QAOA_cct/qaoa_experiments/data"

# directory_path = "../data"
cycle_num = 13
random_shots = 10

# Initialize a plot
plt.figure()


gqra_circuit_shots_scores = []

for i, num_qubits in enumerate(circuit_sizes):
    print("circuit", i+1, "running")

    n = safe_eval(num_qubits)
    # random_shots = 10*np.log(n)
    # random_shots = np.log(n)
    shot_list = [n, n/2, n/4, np.log(n), np.log(n/2), np.log(n/4)]
    gidnet_circuit_list = []

    for shots in shot_list:
        qiskit_circuit, _ = create_qiskit_and_qnet_GRCS_circuits(num_qubits, cycle_num, directory_path)
        random_shots = 10*shots
        gidnet_circuits = gqra_sets(qiskit_circuit, shots=random_shots)
        # gidnet_circuits = gqra_sets(qiskit_circuit, shots=shots)
    
        min_circuits = find_minimum_width_sets(gidnet_circuits)
        min_width, prob = len(min_circuits[0]), len(min_circuits)/random_shots

        # Data colation
        gidnet_circuit_list.append((min_width, prob))

    scores = score_elements(gidnet_circuit_list)
    gqra_circuit_shots_scores.append(scores)

    # Convert shot_list to strings for consistent x-axis representation
    shot_labels = [f"{int(s)}" if s >= 1 else f"{s:.2f}" for s in shot_list]
    
    # Bar chart plotting
    x_labels = ['n', 'n/2', 'n/4', 'log(n)', 'log(n/2)', 'log(n/4)']
    x = np.arange(len(x_labels))  # the label locations
    
    plt.bar(x + i * 0.15, scores, width=0.15, label=f'Circuit Size {num_qubits}')

# Finalize the plot
plt.title('Scores vs Shots for Different Circuit Sizes')
plt.xlabel('Shots')
plt.ylabel('Scores')
plt.xticks(x + 0.15, x_labels)  # Set x-ticks in the center of grouped bars
# plt.grid(True)
plt.legend()
plt.show()

#     # Plotting scores against shot_list in the same plot
#     plt.plot(shot_list, scores, marker='o', linestyle='-', label=f'Circuit Size {num_qubits}')

# # Finalize the plot
# plt.title('Scores vs Shots for Different Circuit Sizes')
# plt.xlabel('Shots')
# plt.ylabel('Scores')
# plt.xscale('log')  # Assuming shots can be on a logarithmic scale
# plt.yscale('log')  # Assuming scores can be on a logarithmic scale
# plt.grid(True)
# plt.legend()
# plt.show()
```
