```python
import matplotlib.pyplot as plt
import ast

# Load the data from the uploaded CSV file
file_path = 'data_qaoa_cct/combined_width_and_runtime_qaoa_data.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe and the columns to understand its structure
data.head(), data.columns

# Convert the string representations of lists into actual lists
data['GidNET Width'] = data['GidNET Width'].apply(ast.literal_eval)
data['QNET Width'] = data['QNET Width'].apply(ast.literal_eval)
data['Qiskit Width'] = data['Qiskit Width'].apply(ast.literal_eval)

# Prepare data for the box plot
gidnet_widths = [width for sublist in data['GidNET Width'] for width in sublist]
qnet_widths = [width for sublist in data['QNET Width'] for width in sublist]
qiskit_widths = [width for sublist in data['Qiskit Width'] for width in sublist]
circuit_sizes_expanded = [[size] * len(widths) for size, widths in zip(data['Circuit Sizes'], data['GidNET Width'])]
circuit_sizes_expanded = [item for sublist in circuit_sizes_expanded for item in sublist]




# Create separate lists for each unique circuit size for each width category
gidnet_widths_by_size = {size: [] for size in data['Circuit Sizes'].unique()}
qnet_widths_by_size = {size: [] for size in data['Circuit Sizes'].unique()}
qiskit_widths_by_size = {size: [] for size in data['Circuit Sizes'].unique()}

# Append widths to the respective lists based on the circuit size
for index, row in data.iterrows():
    gidnet_widths_by_size[row['Circuit Sizes']].extend(row['GidNET Width'])
    qnet_widths_by_size[row['Circuit Sizes']].extend(row['QNET Width'])
    qiskit_widths_by_size[row['Circuit Sizes']].extend(row['Qiskit Width'])

# Plotting the box plots
circuit_sizes_sorted = sorted(gidnet_widths_by_size.keys())



# Prepare data for combined box plot with legends
fig, ax = plt.subplots(figsize=(12, 8))

# Positions for the boxes
positions_gidnet = [x - 0.2 for x in range(len(circuit_sizes_sorted))]
positions_qnet = [x for x in range(len(circuit_sizes_sorted))]
positions_qiskit = [x + 0.2 for x in range(len(circuit_sizes_sorted))]

# GidNET Widths
box_gidnet = ax.boxplot([gidnet_widths_by_size[size] for size in circuit_sizes_sorted], positions=positions_gidnet, widths=0.2, patch_artist=True, boxprops=dict(facecolor='lightgreen'), labels=circuit_sizes_sorted)
# QNET Widths
box_qnet = ax.boxplot([qnet_widths_by_size[size] for size in circuit_sizes_sorted], positions=positions_qnet, widths=0.2, patch_artist=True, boxprops=dict(facecolor='lightblue'))
# Qiskit Widths
box_qiskit = ax.boxplot([qiskit_widths_by_size[size] for size in circuit_sizes_sorted], positions=positions_qiskit, widths=0.2, patch_artist=True, boxprops=dict(facecolor='salmon'))

# Adding legend
ax.legend([box_gidnet["boxes"][0], box_qnet["boxes"][0], box_qiskit["boxes"][0]], ['GidNET Widths', 'QNET Widths', 'Qiskit Widths'], loc='upper left')

ax.set_xlabel('Initial Circuit Size')
ax.set_ylabel('Final Circuit Size')
ax.set_title('Comparison of GidNET, QNET, and Qiskit Qubit Reuse Algorithms on U3R Max-cut QAOA Circuits, p=1')
plt.xticks(range(len(circuit_sizes_sorted)), labels=circuit_sizes_sorted)  # ensure x-ticks are labeled with circuit sizes

# Save figure to a file
fig_path = 'plots_qaoa_cct/GidNET_QNET_and_Qiskit_qaoa_initial_vs_final_circuit_sizes_boxplot.pdf'
plt.savefig(fig_path)

plt.show()
```

### Percentage Speedup

```python
import pandas as pd

# Load the processed data
file_path = 'data_qaoa_cct/processed_GidNET_QNET_and_Qiskit_width_and_runtime_qaoa_data_complete.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe and the column names to understand its structure
data.head(), data.columns


# Calculate the average speed improvement of GidNET over QNET and Qiskit
speed_improvement_over_qnet = (data['QNET Average Runtime'] - data['GidNET Average Runtime']) / data['QNET Average Runtime'] * 100
speed_improvement_over_qiskit = (data['Qiskit Average Runtime'] - data['GidNET Average Runtime']) / data['Qiskit Average Runtime'] * 100

# Calculate overall average improvements
avg_speed_improvement_over_qnet = speed_improvement_over_qnet.mean()
avg_speed_improvement_over_qiskit = speed_improvement_over_qiskit.mean()

avg_speed_improvement_over_qnet, avg_speed_improvement_over_qiskit
```

```python
import pandas as pd

# Load the processed data file
processed_data_path = "data_qaoa_cct/processed_GidNET_QNET_and_Qiskit_width_and_runtime_qaoa_data_complete.csv"
processed_data = pd.read_csv(processed_data_path)

# Display the first few rows of the dataframe and its columns to understand its structure
processed_data.head(), processed_data.columns


# Calculate the ratio of GidNET's average runtime to QNET's and Qiskit's average runtimes
processed_data['GidNET vs QNET Speedup'] = processed_data['QNET Average Runtime'] / processed_data['GidNET Average Runtime']
processed_data['GidNET vs Qiskit Speedup'] = processed_data['Qiskit Average Runtime'] / processed_data['GidNET Average Runtime']

# Extract relevant columns for comparison
comparison_data = processed_data[['Circuit Sizes', 'GidNET Average Runtime', 'QNET Average Runtime', 'Qiskit Average Runtime', 
                                  'GidNET vs QNET Speedup', 'GidNET vs Qiskit Speedup']]

# Calculate the average speedup of GidNET over QNET and Qiskit
average_speedup = {
    'GidNET vs QNET Average Speedup': comparison_data['GidNET vs QNET Speedup'].mean(),
    'GidNET vs Qiskit Average Speedup': comparison_data['GidNET vs Qiskit Speedup'].mean()
}

comparison_data, average_speedup

```
