```python
import pandas as pd

# Assuming you've already loaded these DataFrames or you can load them from CSVs directly
df_gidnet = pd.read_csv('data_supremacy_cct/GidNET_width_and_runtime_supremacy_cct_data.csv')
df_qnet = pd.read_csv('data_supremacy_cct/QNET_width_and_runtime_supremacy_cct_data.csv')
# df_qnet = pd.read_csv('data_supremacy_cct/QNET-main_width_and_runtime_supremacy_cct_data.csv')
df_qiskit = pd.read_csv('data_supremacy_cct/Qiskit_width_and_runtime_supremacy_cct_data.csv')

# Merging the DataFrames on 'Circuit Sizes'
df_combined = pd.merge(df_gidnet, df_qnet, on='Circuit Sizes', how='outer', suffixes=('_gidnet', '_qnet'))
df_combined = pd.merge(df_combined, df_qiskit, on='Circuit Sizes', how='outer', suffixes=('', '_qiskit'))

# Save the combined DataFrame to CSV
df_combined.to_csv('data_supremacy_cct/combined_width_and_runtime_supremacy_data.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load the data
data_path = 'data_supremacy_cct/combined_width_and_runtime_supremacy_data.csv'
df = pd.read_csv(data_path)

# Plotting the timing data
plt.figure(figsize=(10, 6))

# Plot for deterministic method
plt.plot(df["Circuit Sizes"], df["GidNET Width"], label='GidNET', marker='o', color="green")

plt.plot(df["Circuit Sizes"], df["QNET Width"], label='QNET', marker='^', color="blue")
# plt.plot(df["Circuit Sizes"], df["QNET-main Width"], label='QNET', marker='^', color="blue")

# Plot for random method
plt.plot(df["Circuit Sizes"], df["Qiskit Width"], label='Qiskit', marker='x', color="red")

# Adding titles and labels
plt.title('Supremacy Circuit Qubit Reuse Compilation for GidNET, QNET and Qiskit')
plt.xlabel('Iniital Circuit Sizes')
plt.ylabel('Final Circuit Sizes')
plt.legend()

# Show grid
# plt.grid(True)

# Save figure to a file
# fig_path = 'QNET_runtime.png'
fig_path = 'plots_supremacy_cct/GidNET_QNET_and_Qiskit_initial_vs_final_supremacy_circuit_sizes.pdf'
plt.savefig(fig_path)

# Display the plot
plt.show()

```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load the data
data_path = 'data_supremacy_cct/combined_width_and_runtime_supremacy_data.csv'
df = pd.read_csv(data_path)

# Setting up the figure
plt.figure(figsize=(12, 8))

# Defining the width of a bar and the positions
bar_width = 0.25
positions = np.arange(len(df['Circuit Sizes']))

# GidNET
# plt.bar(positions - bar_width, df["GidNET Width"], yerr=df["GidNET Stdev Width"], width=bar_width, label='GidNET', color="blue", capsize=5)
plt.bar(positions - bar_width, df["GidNET Width"], width=bar_width, label='GidNET', color="green", capsize=5)

# GidNET+
plt.bar(positions, df["QNET Width"], width=bar_width, label='QNET', color="blue", capsize=5)
# plt.bar(positions, df["QNET-main Width"], width=bar_width, label='QNET', color="blue", capsize=5)

# QNET
plt.bar(positions + bar_width, df["Qiskit Width"], width=bar_width, label='Qiskit', color="red", capsize=5)

# Setting the position of x-ticks to the middle of the group bars
plt.xticks(positions, df["Circuit Sizes"])

# Adding titles and labels
plt.title('Supremacy Circuit Qubit Reuse Compilation for GidNET, QNET and Qiskit')
plt.xlabel('Initial Circuit Sizes')
plt.ylabel('Final Circuit Sizes')
plt.legend()

# Show grid
# plt.grid(True)

# Save figure to a file
fig_path = 'plots_supremacy_cct/GidNET_QNET_and_Qiskit_initial_vs_final_supremacy_circuit_width_bars.pdf'
plt.savefig(fig_path)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load the data
data_path = 'data_supremacy_cct/combined_width_and_runtime_supremacy_data.csv'
df = pd.read_csv(data_path)

# Plotting the timing data
plt.figure(figsize=(10, 6))

# Plot for deterministic method
plt.plot(df["Circuit Sizes"], df["GidNET Average Runtime"], label='GidNET', marker='o', color="green")

plt.plot(df["Circuit Sizes"], df["QNET Average Runtime"], label='QNET', marker='^', color="blue")
# plt.plot(df["Circuit Sizes"], df["QNET-main Average Runtime"], label='QNET', marker='^', color="blue")

# Plot for random method
plt.plot(df["Circuit Sizes"], df["Qiskit Average Runtime"], label='Qiskit', marker='x', color="red")

# Adding titles and labels
plt.title('Supremacy Circuit Qubit Reuse Compilation for GidNET, QNET and Qiskit')
plt.xlabel('Iniital Circuit Sizes')
plt.ylabel('Average Runtime (s)')
plt.legend()

plt.xticks(fontweight='bold', fontsize=10)
plt.yticks(fontweight='bold', fontsize=10)

# Show grid
# plt.grid(True)

# Save figure to a file
# fig_path = 'QNET_runtime.png'
fig_path = 'plots_supremacy_cct/GidNET_QNET_and_Qiskit_initial_supremacy_circuit_sizes_vs_average_runtime.pdf'
plt.savefig(fig_path)

# Display the plot
plt.show()

```
