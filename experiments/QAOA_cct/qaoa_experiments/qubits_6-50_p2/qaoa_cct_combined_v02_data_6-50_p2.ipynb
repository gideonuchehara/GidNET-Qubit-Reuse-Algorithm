```python
import matplotlib.pyplot as plt
import pandas as pd
import ast

# Load the data from the uploaded CSV file
file_path = 'data_qaoa_cct/combined_v02_width_and_runtime_qaoa_p2_data_complete.csv'
data = pd.read_csv(file_path)

# Convert the string representations of lists into actual lists
data['GidNET Width'] = data['GidNET Width'].apply(ast.literal_eval)
data['QNET Width'] = data['QNET Width'].apply(ast.literal_eval)
# data['Qiskit Width'] = data['Qiskit Width'].apply(ast.literal_eval)

# Prepare data for the box plot
gidnet_widths = [width for sublist in data['GidNET Width'] for width in sublist]
qnet_widths = [width for sublist in data['QNET Width'] for width in sublist]
# qiskit_widths = [width for sublist in data['Qiskit Width'] for width in sublist]
circuit_sizes_expanded = [[size] * len(widths) for size, widths in zip(data['Circuit Sizes'], data['GidNET Width'])]
circuit_sizes_expanded = [item for sublist in circuit_sizes_expanded for item in sublist]

# Create separate lists for each unique circuit size for each width category
gidnet_widths_by_size = {size: [] for size in data['Circuit Sizes'].unique()}
qnet_widths_by_size = {size: [] for size in data['Circuit Sizes'].unique()}
# qiskit_widths_by_size = {size: [] for size in data['Circuit Sizes'].unique()}

# Append widths to the respective lists based on the circuit size
for index, row in data.iterrows():
    gidnet_widths_by_size[row['Circuit Sizes']].extend(row['GidNET Width'])
    qnet_widths_by_size[row['Circuit Sizes']].extend(row['QNET Width'])
    # qiskit_widths_by_size[row['Circuit Sizes']].extend(row['Qiskit Width'])

# Plotting the box plots
circuit_sizes_sorted = sorted(gidnet_widths_by_size.keys())

# Prepare data for combined box plot with legends
fig, ax = plt.subplots(figsize=(8, 6), dpi=400)  # Adjusted size to better suit typical LaTeX document width

# Positions for the boxes
positions_gidnet = [x - 0.2 for x in range(len(circuit_sizes_sorted))]
positions_qnet = [x for x in range(len(circuit_sizes_sorted))]
# positions_qiskit = [x + 0.2 for x in range(len(circuit_sizes_sorted))]

# GidNET Widths
box_gidnet = ax.boxplot([gidnet_widths_by_size[size] for size in circuit_sizes_sorted], positions=positions_gidnet, widths=0.2, patch_artist=True, boxprops=dict(facecolor='lightgreen'), labels=circuit_sizes_sorted)
# QNET Widths
box_qnet = ax.boxplot([qnet_widths_by_size[size] for size in circuit_sizes_sorted], positions=positions_qnet, widths=0.2, patch_artist=True, boxprops=dict(facecolor='lightblue'))
# Qiskit Widths
# box_qiskit = ax.boxplot([qiskit_widths_by_size[size] for size in circuit_sizes_sorted], positions=positions_qiskit, widths=0.2, patch_artist=True, boxprops=dict(facecolor='salmon'))

# Adding legend with bold fonts
ax.legend([box_gidnet["boxes"][0], box_qnet["boxes"][0]], ['GidNET', 'QNET'], loc='upper left', prop={'weight': 'bold'})

# Setting labels with bold fonts
ax.set_xlabel('Initial Circuit Width', fontsize=14, fontweight='bold')
ax.set_ylabel('Final Circuit width', fontsize=14, fontweight='bold')
# ax.set_title('Comparison of GidNET, QNET, and Qiskit Qubit Reuse \n Algorithms on U3R Max-cut QAOA Circuits, p=1', fontsize=14, fontweight='bold')
plt.xticks(range(len(circuit_sizes_sorted)), labels=circuit_sizes_sorted, fontweight='bold', fontsize=14)  # Bold and size-adjusted tick labels
plt.yticks(fontweight='bold', fontsize=14)  # Bold and size-adjusted tick labels

# Save figure to a file
fig_path = 'plots_qaoa_cct/GidNET_v02_and_QNET_and_qaoa_p2_initial_vs_final_circuit_width_boxplot.pdf'
plt.savefig(fig_path, format='pdf', bbox_inches='tight')

# Display the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Define your circuit sizes and corresponding theoretical and experimental values
circuit_size = np.array(df["Circuit Sizes"])
theoretical_values = np.array(model_gidnet(df["Circuit Sizes"], *popt_gidnet))
experimental_values = np.array(df["GidNET Average Runtime"])

# Fit the polynomial regression
degree=2
poly_features = PolynomialFeatures(degree)
X_poly = poly_features.fit_transform(circuit_size.reshape(-1, 1))
model = LinearRegression().fit(X_poly, experimental_values)
predicted_values = model.predict(X_poly)

# Calculate R-squared value
r_squared = r2_score(experimental_values, predicted_values)

# Compute the F-statistic
n = len(circuit_size)
k = degree + 1  # Degree of polynomial + 1 (for the intercept)
f_statistic = (r_squared / (1 - r_squared)) * ((n - k - 1) / k)

print("R-squared value:", r_squared)
print("F-statistic:", f_statistic)
```

```python
import pandas as pd
import numpy as np

# Load the data
data_path = "data_qaoa_cct/processed_GidNET_v02_QNET_and_Qiskit_width_and_runtime_qaoa_p1_data_complete.csv"
data = pd.read_csv(data_path)

# Calculate the individual speedups
speedup_vs_qnet = data['QNET Average Runtime'] / data['GidNET Average Runtime']
# speedup_vs_qiskit = data['Qiskit Average Runtime'] / data['GidNET Average Runtime']

# Calculate the geometric mean of the speedups
geo_mean_speedup_vs_qnet = np.prod(speedup_vs_qnet) ** (1 / len(speedup_vs_qnet))
# geo_mean_speedup_vs_qiskit = np.prod(speedup_vs_qiskit) ** (1 / len(speedup_vs_qiskit))

# Calculate the percentage execution time reduction
percent_reduction_vs_qnet = 100 * (1 - 1 / geo_mean_speedup_vs_qnet)
# percent_reduction_vs_qiskit = 100 * (1 - 1 / geo_mean_speedup_vs_qiskit)

# Create a summary DataFrame
performance_improvement_geo_mean = pd.DataFrame({
    'Geometric Mean Speedup vs QNET (X)': [geo_mean_speedup_vs_qnet],
    'Percentage Execution Time Reduction vs QNET (%)': [percent_reduction_vs_qnet],
    # 'Geometric Mean Speedup vs Qiskit (X)': [geo_mean_speedup_vs_qiskit],
    # 'Percentage Execution Time Reduction vs Qiskit (%)': [percent_reduction_vs_qiskit]
})

performance_improvement_geo_mean
```
