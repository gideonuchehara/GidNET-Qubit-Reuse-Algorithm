```python
from gidnet_qubit_reuse_algorithm.gidnet_qra import (modified_compute_qubit_reuse_sets
                                                    )

from helper_functions.benchmarks import generate_circ
from qiskit import QuantumCircuit
import numpy as np
```

```python
circuit_sizes = [16, 20, 25, 30, 36, 42, 49, 56, 64, 72, 81, 90, 100, 110, 121, 132, 144]
random_shots = 10 # shots for qnet algorithm

gidnet_average_time = []
gidnet_std_time = []
gidnet_qubit_width = []

for i, circuit_size in enumerate(circuit_sizes):
    print("circuit", i+1, "running")
    circuit_type = "supremacy"

    supremacy_circuit = generate_circ(
            num_qubits=circuit_size,
            depth=1,
            circuit_type=circuit_type,
            reg_name="q",
            connected_only=True,
            seed=None,
    )

    if supremacy_circuit is not None:
        supremacy_circuit.measure_all()


    #### GIDNET EXPERIMENT BEGINS #####
    gidnet_time = %timeit -o modified_compute_qubit_reuse_sets(supremacy_circuit, shots=random_shots)
    gidnet_result = modified_compute_qubit_reuse_sets(supremacy_circuit, shots=random_shots)

    gidnet_average_time.append(gidnet_time.average)
    gidnet_std_time.append(gidnet_time.stdev)
    gidnet_qubit_width.append(len(gidnet_result))
```

```python
import pandas as pd
# Convert lists to a dictionary
data = {
    'Circuit Sizes': circuit_sizes,
    'GidNET Average Runtime': gidnet_average_time,  # Added comma
    'GidNET Stdev Runtime': gidnet_std_time,  # Added comma
    'GidNET Width': gidnet_qubit_width,
}

# Assuming pandas is imported as pd
import pandas as pd

# Convert dictionary to DataFrame
df = pd.DataFrame(data)

# Save DataFrame to CSV
df.to_csv('GidNET_v03_width_and_runtime_supremacy_cct_data.csv', index=False)
```
